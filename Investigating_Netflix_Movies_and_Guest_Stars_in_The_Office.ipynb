```python
data = data[["id",'title', 'genre','overview']]
```

```python
data.head()
```

```python
data["tag"] =data["genre"] + data["overview"]
```

```python
data.head()
```

```python
data.drop(columns=["genre","overview"])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000,stop_words="english")
```

```python
vector = cv.fit_transform(data["tag"].values.astype("U")).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
data[data["title"]=="The Godfather"].index[0]
```

```python
similarity[2]
```

```python
distance = sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
```

```python
distance[:5]
```

```python
for i in distance[0:5]:
  print(data.iloc[i[0]].title)
```

```python
def recommand(movies):
  index = data[data["title"]==movies].index[0]
  distance = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
  for i in distance[0:5]:
    print(data.iloc[i[0]].title)
```

```python
recommand("Iron Man")
```
